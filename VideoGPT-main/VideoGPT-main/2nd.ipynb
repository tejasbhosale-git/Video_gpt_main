```python
import cv2
import os

pipe_path = 'my_pipe'

if not os.path.exists(pipe_path):
    os.mkfifo(pipe_path)

cap = cv2.VideoCapture(0)

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        with open(pipe_path, 'wb') as f:
            f.write(frame.tobytes())
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
