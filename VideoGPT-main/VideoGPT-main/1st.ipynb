```python
import re
import os
def sort_files_numerically(folder_path):
    files = os.listdir(folder_path)
    files = [f for f in files if re.match(r'clips_\d+\.mp4', f)]
    files.sort(key=lambda x: int(re.search(r'(\d+)', x).group(1)))
    return files

folder_path = '/Users/krisanusarkar/Documents/ML/videototext/educlips'

sorted_files = sort_files_numerically(folder_path)
print("Sorted files:", sorted_files)
```

```python
import os
import subprocess

# List of sorted video files
video_files = sorted_files

# Check if all video files exist
for video in video_files:
    if not os.path.isfile(video):
        raise FileNotFoundError(f"File not found: {video}")

# Create a text file for the concat protocol
with open('input_list.txt', 'w') as f:
    for video in video_files:
        f.write(f"file '{os.path.abspath(video)}'\n")

# Construct the FFmpeg command
command = 'ffmpeg -f concat -safe 0 -i input_list.txt -c:v libx264 -c:a aac output2.mp4'

# Print the command for debugging purposes
print("Generated FFmpeg command:")
print(command)

```

```python
import re
def find_just_greater(sorted_array, number):
    left, right = 0, len(sorted_array) - 1
    result = None
    
    while left <= right:
        mid = (left + right) // 2
        if sorted_array[mid] > number:
            result = sorted_array[mid]
            right = mid - 1
        else:
            left = mid + 1
    
    return result


def conversionoftimeframe(number):
    result = find_just_greater(list(timeframe_dict.keys()), number)
    if result is None :
        return list(timeframe_dict.values())[-1]
    return round(timeframe_dict[result] + number - result , 2)

def extract_mm_ss_timestamps(text):
    # Regex pattern for MM:SS format
    pattern = r'\b[0-5]?[0-9]:[0-5][0-9]\b'

    # Find all matches
    matches = re.findall(pattern, text)

    return matches


def seconds_to_mm_ss(total_seconds):
    minutes = int(total_seconds // 60)
    seconds = int(total_seconds % 60)
    return f"{minutes}:{seconds:02d}"

def mm_ss_to_seconds(mm_ss_timestamp):
    minutes, seconds = map(int, mm_ss_timestamp.split(':'))
    total_seconds = minutes * 60 + seconds
    return total_seconds

def convert_text(text):
    timestamps = extract_mm_ss_timestamps(text)
    for timestamp in timestamps:
        text = text.replace(timestamp , seconds_to_mm_ss(conversionoftimeframe(mm_ss_to_seconds(timestamp))))
```
